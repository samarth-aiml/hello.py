```python
#if else elif logic
#simple grader checker via if else logic
student = str(input("Enter the student's name: "))
grade = float(input(f"Enter Grade of {student} :"))
#logic of grading system
if(grade >= 90):
  print("A+ Grade")
elif(grade >= 80 and grade > 70):
  print("B Grade")
elif(grade >= 60 and grade >= 50):
  print("C Grade")
else:
  print('Failed')
```
