```python
# Simple Miles to Kilometers Converter Project

print("--- Miles to Kilometers Converter ---")

# 1. Get input from the user for the distance in miles with error handling
while True:
    try:
        miles_str = input("Enter distance in miles: ")
        miles = float(miles_str)
        break # Exit loop if input is valid
    except ValueError:
        print("Invalid input. Please enter a numerical value for miles.")

# 2. Perform the conversion calculation
# 1 mile is approximately 1.60934 kilometers
kilometers = miles * 1.60934

# 3. Display the result using the .format() method (no f-strings)
print("{} miles is equal to {} kilometers.".format(miles, kilometers))

print("----------------------------------------")
print("Project complete! Run it with different distances.")
```
